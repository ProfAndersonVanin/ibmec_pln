# Aplicação 1 — Análise de Sentimentos

**Objetivo**: Classificar textos como Positivo ou Negativo

````python
from transformers import pipeline

classificador = pipeline("sentiment-analysis")

resultado = classificador("Este curso de PLN é excelente!")

print(resultado)
````

Teste outros exemplos

````python
frases = [
    "O filme foi incrível",
    "Não gostei da prova",
    "A aula está muito boa",
    "O atendimento foi péssimo"
]

for frase in frases:
    print(classificador(frase))
````

# Aplicação 2 — Geração de Texto
**Objetivo**: Gerar automaticamente continuações de frases.

````python
from transformers import pipeline

gerador = pipeline("text-generation")

texto = gerador(
    "A inteligência artificial irá",
    max_length=50,
    num_return_sequences=1
)

print(texto)
````

**Explicação**

Parâmetros importantes:

- max_length --> tamanho máximo do texto
- num_return_sequences --> quantidade de respostas
- temperature --> criatividade do modelo

Exemplo com temperatura

````python
texto = gerador(
    "No futuro da educação",
    max_length=60,
    temperature=0.9
)

print(texto)
````

# Aplicação 3 — Perguntas e Respostas
**Objetivo**: O modelo responde perguntas com base em um contexto.

````python
from transformers import pipeline

qa = pipeline("question-answering")

contexto = """
Python é uma linguagem de programação muito utilizada em
Ciência de Dados, Inteligência Artificial e Desenvolvimento Web.
"""

pergunta = "Onde Python é utilizado?"

resultado = qa(
    question=pergunta,
    context=contexto
)

print(resultado)
````

# Aplicação 4 — Sumarização de Texto
**Objetivo**: Gerar resumos automáticos.

````python
from transformers import pipeline

sumarizador = pipeline(
    "text-generation",
    model="google-t5/t5-small"
)

texto = """
A Inteligência Artificial vem crescendo rapidamente nos últimos anos.
Diversas empresas estão investindo em soluções automatizadas para análise
de dados, atendimento ao cliente e automação de processos.
O uso de modelos de linguagem natural também aumentou bastante.
"""

entrada = "summarize: " + texto

resumo = sumarizador(
    entrada,
    max_length=40,
    min_length=10,
    do_sample=False
)

print(resumo)
````

# Aplicação 5 — Classificação Zero-Shot
**Objetivo**: Classificar textos sem treinar um modelo.

````python
from transformers import pipeline

classificador = pipeline("zero-shot-classification")

texto = "O novo processador possui maior desempenho gráfico."

labels = [
    "tecnologia",
    "esportes",
    "política",
    "educação"
]

resultado = classificador(texto, labels)

print(resultado)
````

# Trabalhando com Tokenizers

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize("Transformers são incríveis!")

print(tokens)
````

# Convertendo Tokens em IDs

````python
ids = tokenizer.convert_tokens_to_ids(tokens)

print(ids)
````

# Exemplo com Modelo BERT

````python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
modelo = AutoModel.from_pretrained("bert-base-uncased")

texto = "Aprendendo PLN com Transformers"

inputs = tokenizer(texto, return_tensors="pt")

outputs = modelo(**inputs)

print(outputs.last_hidden_state.shape)
````
