## Sobre o BERTimbau

O BERTimbau é uma versão do modelo BERT treinada especificamente para o idioma português.

Ele foi desenvolvido para tarefas de Processamento de Linguagem Natural (PLN), como:

- Classificação de texto
- Análise de sentimentos
- Respostas automáticas
- Reconhecimento de entidades
- Sumarização
- Tradução

Nesta aula vamos utilizar o modelo para **Análise de Sentimentos em Português**.

# Etapa 2 - Importação das Bibliotecas

Agora vamos importar as bibliotecas necessárias.


````python
from transformers import pipeline
````

# Etapa 3 - Carregando o Modelo BERTimbau

Nesta etapa vamos carregar um modelo já treinado para análise de sentimentos.

O pipeline simplifica bastante o uso de modelos de PLN.


````python
modelo = pipeline(
    "sentiment-analysis",
    model="neuralmind/bert-base-portuguese-cased",
    tokenizer="neuralmind/bert-base-portuguese-cased"
)

print("Modelo carregado com sucesso!")
````

# Observação Importante

O modelo original do BERTimbau não foi treinado especificamente para sentimentos.

Por isso, em aplicações reais normalmente utilizamos versões ajustadas (fine-tuning).

Mesmo assim, esta aula é excelente para compreender:
- Como carregar modelos
- Como tokenizar textos
- Como realizar inferência
- Como utilizar modelos pré-treinados


# Etapa 4 - Criando uma Função para Análise de Texto

Separar funcionalidades em funções deixa o código mais organizado e reutilizável.


````python
def analisar_sentimento(texto):
    resultado = modelo(texto)
    return resultado
````

# Etapa 5 - Testando Frases Positivas


````python
frase = "Este curso de inteligência artificial é excelente!"

resultado = analisar_sentimento(frase)

print("Texto:", frase)
print("Resultado:", resultado)
````

# Etapa 6 - Testando Frases Negativas


````python
frase = "O atendimento foi muito ruim e demorado."

resultado = analisar_sentimento(frase)

print("Texto:", frase)
print("Resultado:", resultado)
````

# Etapa 7 - Testando Diversas Frases

Agora vamos analisar várias frases de uma única vez.


````python
frases = [
    "Gostei muito deste filme.",
    "O produto chegou quebrado.",
    "A aula foi muito interessante.",
    "Estou decepcionado com o serviço.",
    "Esse restaurante é maravilhoso."
]

for frase in frases:
    resultado = analisar_sentimento(frase)

    print("-" * 50)
    print("Frase:", frase)
    print("Resultado:", resultado)
````

# Etapa 8 - Entendendo o Resultado

O modelo normalmente retorna:

- LABEL: classe prevista
- SCORE: nível de confiança da previsão

Exemplo:

```python
[{'label': 'LABEL_1', 'score': 0.98}]
```

Quanto mais próximo de 1, maior a confiança do modelo.


# Etapa 9 - Criando uma Interface Mais Amigável

Vamos melhorar a saída do modelo.


````python
def interpretar_resultado(resultado):

    label = resultado[0]['label']
    score = resultado[0]['score']

    print(f"Classe prevista: {label}")
    print(f"Confiança: {score:.4f}")
````

````python
frase = "A experiência foi fantástica!"

resultado = analisar_sentimento(frase)

print("Texto:", frase)
interpretar_resultado(resultado)
````

# Etapa 10 - Utilizando um Modelo Ajustado para Sentimentos

Agora vamos utilizar um modelo treinado especificamente para análise de sentimentos em Português.


````python
modelo_sentimentos = pipeline(
    "sentiment-analysis",
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
)

print("Modelo de sentimentos carregado!")
````

# Etapa 11 - Nova Função de Predição


````python
def prever_sentimento(texto):

    resultado = modelo_sentimentos(texto)

    return resultado
````

# Etapa 12 - Testando o Novo Modelo


````python
frases = [
    "Estou muito feliz com o resultado.",
    "Esse notebook ficou incrível.",
    "Não gostei da qualidade do produto.",
    "A entrega atrasou bastante."
]

for frase in frases:

    resultado = prever_sentimento(frase)

    print("-" * 50)
    print("Texto:", frase)
    print(resultado)
````

# Etapa 13 - Explicação Conceitual

## Como o BERTimbau funciona?

O modelo transforma palavras em vetores numéricos.

Esses vetores representam:
- significado
- contexto
- relação entre palavras

Diferente de modelos antigos, o BERT consegue analisar:
- palavras anteriores
- palavras posteriores

Isso melhora bastante a compreensão do texto.

## Exemplo

Na frase:

```text
O banco estava lotado.
```

A palavra "banco" pode significar:
- instituição financeira
- assento

O contexto ajuda o modelo a entender o significado correto.
