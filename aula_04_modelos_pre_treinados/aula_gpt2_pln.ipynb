
# Etapa 2 - Importando as Bibliotecas


````python

from transformers import pipeline
````


# Etapa 3 - Carregando o Modelo GPT-2

Vamos utilizar um pipeline de geração de texto.


````python

gerador = pipeline(
    "text-generation",
    model="gpt2"
)

print("Modelo carregado com sucesso!")
````


# Etapa 4 - Gerando o Primeiro Texto

O modelo recebe um texto inicial chamado:

- prompt
- contexto inicial
- entrada

E tenta continuar automaticamente.


````python

texto = "A inteligência artificial"

resultado = gerador(
    texto,
    max_length=50,
    num_return_sequences=1
)

print(resultado[0]['generated_text'])
````

> Para trabalhar com português, use um modelo multilíngue ou treinado em português.



````python
gerador = pipeline(
    "text-generation",
    model="pierreguillou/gpt2-small-portuguese"
)

texto = "A inteligência artificial"

resultado = gerador(
    texto,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.8
)

print(resultado[0]['generated_text'])
````


# Etapa 5 - Criando uma Função de Geração

Separar o código em funções deixa o notebook mais organizado.


````python

def gerar_texto(prompt, tamanho=50):

    resultado = gerador(
        prompt,
        max_length=tamanho,
        num_return_sequences=1
    )

    return resultado[0]['generated_text']
````


# Etapa 6 - Testando Diferentes Prompts


````python

prompts = [
    "O futuro da robótica",
    "Python é uma linguagem",
    "As redes neurais",
    "A visão computacional"
]

for prompt in prompts:

    texto = gerar_texto(prompt)

    print("-" * 50)
    print(texto)
    print()
````


# Etapa 7 - Controlando o Tamanho do Texto

O parâmetro `max_length` controla o tamanho máximo da saída.


````python

texto = gerar_texto(
    "A tecnologia moderna",
    tamanho=100
)

print(texto)
````


# Etapa 8 - Controlando Criatividade

Podemos ajustar:

- temperature
- top_k
- top_p

Esses parâmetros influenciam a criatividade do modelo.


````python

def gerar_texto_criativo(prompt):

    resultado = gerador(
        prompt,
        max_length=80,
        temperature=0.9,
        top_k=50,
        top_p=0.95,
        do_sample=True
    )

    return resultado[0]['generated_text']
````


# Etapa 9 - Gerando Texto Criativo


````python

texto = gerar_texto_criativo(
    "No ano de 2050 os robôs"
)

print(texto)
````


# Etapa 10 - Explicando os Parâmetros

## temperature

Controla a criatividade.

- baixo valor → respostas mais previsíveis
- alto valor → respostas mais criativas

## top_k

Limita quantas palavras podem ser escolhidas.

## top_p

Seleciona palavras com maior probabilidade acumulada.



# Etapa 11 - Criando um Pequeno Chat

Agora vamos criar uma interface simples.


````python

while True:

    entrada = input("Digite um texto (ou sair): ")

    if entrada.lower() == "sair":
        break

    resposta = gerar_texto_criativo(entrada)

    print("\nResposta:")
    print(resposta)
    print()
````


# Etapa 12 - Limitações do GPT-2

O GPT-2:

- pode inventar informações
- pode repetir frases
- não entende fatos como humanos
- gera texto baseado em probabilidades

Por isso:
- respostas devem ser revisadas
- não deve ser usado sozinho em aplicações críticas
