
# Exemplo Prático - Modelo de Perguntas e Respostas com BERTimbau

## Objetivo

Nesta atividade vamos criar um sistema simples capaz de responder perguntas em Português utilizando modelos de PLN.

O sistema irá:

- Receber um contexto
- Receber uma pergunta
- Encontrar a resposta automaticamente

Esse tipo de aplicação é conhecido como:

- Question Answering (QA)
- Perguntas e Respostas
- Extração de Respostas em Texto



# Etapa 1 - Instalação das Bibliotecas


# Etapa 2 - Importação das Bibliotecas


````python

from transformers import pipeline
````


# Etapa 3 - Carregando o Modelo

Vamos utilizar um pipeline de Question Answering.

O modelo consegue localizar respostas dentro de um texto.


````python

modelo_qa = pipeline(
    "question-answering",
    model="pierreguillou/bert-base-cased-squad-v1.1-portuguese"
)

print("Modelo carregado com sucesso!")
````


# Etapa 4 - Criando o Texto Base

O modelo precisa de um contexto para procurar respostas.


````python

contexto = '''
A Inteligência Artificial é uma área da computação que busca criar sistemas capazes
de executar tarefas que normalmente exigem inteligência humana.
Essas tarefas incluem reconhecimento de imagens, processamento de linguagem natural,
tomada de decisão e aprendizado automático.

O BERTimbau é um modelo de linguagem treinado especificamente para o idioma português.
Ele pode ser utilizado em tarefas como análise de sentimentos,
classificação de textos e sistemas de perguntas e respostas.
'''
````


# Etapa 5 - Criando uma Função para Perguntas


````python

def responder_pergunta(pergunta, contexto):

    resultado = modelo_qa(
        question=pergunta,
        context=contexto
    )

    return resultado
````


# Etapa 6 - Fazendo a Primeira Pergunta


````python

pergunta = "O que é o BERTimbau?"

resultado = responder_pergunta(pergunta, contexto)

print("Pergunta:", pergunta)
print("Resposta:", resultado['answer'])
print("Confiança:", round(resultado['score'], 4))
````


# Etapa 7 - Testando Novas Perguntas


````python

perguntas = [
    "Para que serve a Inteligência Artificial?",
    "Quais tarefas a IA pode executar?",
    "Para qual idioma o BERTimbau foi treinado?",
    "O que o BERTimbau pode fazer?"
]

for pergunta in perguntas:

    resultado = responder_pergunta(pergunta, contexto)

    print("-" * 50)
    print("Pergunta:", pergunta)
    print("Resposta:", resultado['answer'])
    print("Confiança:", round(resultado['score'], 4))
````


# Etapa 8 - Entendendo o Funcionamento

O modelo funciona da seguinte forma:

1. Recebe um texto (contexto)
2. Recebe uma pergunta
3. Procura dentro do texto o trecho mais provável
4. Retorna a resposta

## Importante

Esse modelo NÃO inventa respostas.

Ele apenas encontra informações já existentes no texto fornecido.



# Etapa 9 - Criando um Sistema Interativo


````python
while True:

    pergunta = input("Digite sua pergunta (ou sair): ")

    if pergunta.lower() == "sair":
        break

    resultado = responder_pergunta(pergunta, contexto)

    print("\nResposta:", resultado['answer'])
    print("Confiança:", round(resultado['score'], 4))
    print()
````
