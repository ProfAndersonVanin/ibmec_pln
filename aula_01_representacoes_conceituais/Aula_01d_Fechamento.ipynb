## 📚 O que estudamos nesta aula?

Iniciamos nossa jornada no PLN respondendo a uma pergunta fundamental:

> **"Como um computador pode entender o significado das palavras?"**

Para responder isso, passamos por quatro grandes conceitos:

---

### 1️⃣ Ontologias

Aprendemos que uma **ontologia** é uma estrutura formal que organiza conceitos  
e as relações entre eles — como um dicionário inteligente que não apenas define  
palavras, mas também sabe que um *cachorro* **é um** *mamífero*, que **tem partes**  
como *pata* e *focinho*, e que é **relacionado** a *lobo* e *raposa*.

Sem ontologias, o computador trata palavras como sequências de letras sem sentido.

---

### 2️⃣ Grafos de Conhecimento

Vimos que um **grafo de conhecimento** é a forma visual de representar uma ontologia,  
usando **nós** (conceitos) e **arestas** (relações). Construímos grafos reais com  
a biblioteca NetworkX, visualizando hierarquias de animais e instrumentos musicais.

Descobrimos que grandes produtos como o **Google**, a **Wikipedia (Wikidata)**  
e a **Alexa** da Amazon são alimentados por grafos de conhecimento.

---

### 3️⃣ WordNet

Exploramos o **WordNet**, o banco de dados léxico mais famoso da língua inglesa,  
criado pela Universidade de Princeton. Aprendemos a:

- Buscar **synsets** (conjuntos de sinônimos que representam um conceito único)
- Navegar pelas relações **hypernym**, **hyponym** e **meronym**
- Percorrer a **hierarquia completa** de um conceito até o topo
- Calcular a **similaridade semântica** entre palavras
- Usar o **WordNet em Português** com o pacote `omw-1.4`

---

### 4️⃣ BabelNet

Conhecemos o **BabelNet**, a evolução multilíngue do WordNet, cobrindo mais de  
500 idiomas ao combinar WordNet, Wikipedia e outros recursos.

Usamos a **API REST** com autenticação por chave (API Key) para:
- Buscar synsets de palavras em Português
- Consultar definições e traduções automáticas
- Explorar relações entre conceitos
- Entender a **ambiguidade lexical** — a palavra *cachorro* retornou  
  **8 synsets diferentes**, cada um representando um significado distinto

## 🌍 Aplicações Práticas no Mundo Real

Os conceitos desta aula não são apenas teóricos —  
eles estão presentes em produtos que você usa todos os dias:

---

### 🔍 1. Mecanismos de Busca (Google, Bing)

Quando você pesquisa **"banco perto de mim"**, o Google usa seu  
**Knowledge Graph** (grafo de conhecimento) para entender que você  
quer uma *instituição financeira*, não um *móvel para sentar*.

Sem ontologias, a busca retornaria resultados irrelevantes.

---

### 🤖 2. Assistentes Virtuais (Alexa, Siri, Google Assistant)

Quando você pergunta *"Alexa, toque uma música do artista que criou Thriller"*,  
o assistente precisa saber que *Michael Jackson* **criou** *Thriller* —  
essa relação está armazenada em um grafo de conhecimento.

---

### 💊 3. Saúde e Medicina

Ontologias médicas como o **SNOMED CT** e o **MeSH** organizam  
doenças, sintomas e medicamentos em hierarquias.  
Sistemas de diagnóstico assistido por IA consultam esses grafos  
para sugerir hipóteses diagnósticas.

---

### 🛒 4. Recomendação de Produtos (Amazon, Mercado Livre)

Quando você pesquisa *"tênis para corrida"*, o sistema de e-commerce  
usa ontologias de produtos para saber que *tênis* é um *calçado*,  
que *corrida* é um *esporte*, e que você provavelmente também se  
interessa por *meias esportivas* e *palmilhas*.

---

### 🌐 5. Tradução Automática (Google Translate, DeepL)

Recursos como o BabelNet alimentam sistemas de tradução,  
fornecendo equivalentes multilíngues de conceitos e ajudando  
o modelo a escolher a tradução correta conforme o contexto.

---

### 📰 6. Análise de Notícias e Fake News

Grafos de conhecimento são usados para verificar se fatos em  
uma notícia são consistentes com o que já se sabe sobre o mundo —  
uma técnica usada em sistemas de detecção de desinformação.

## 📊 Resumo Técnico — O que você aprendeu a usar

````python
# ─────────────────────────────────────────────────────────────
# Resumo executável — demonstração de tudo que foi aprendido
# na Aula 1 em um único bloco de código
# ─────────────────────────────────────────────────────────────

import nltk
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from nltk.corpus import wordnet as wn

nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

print('=' * 60)
print('  RESUMO TÉCNICO — AULA 1')
print('  Representações Conceituais de Palavras')
print('=' * 60)

palavra = 'school'
synsets = wn.synsets(palavra)

print(f'\n🔤 Palavra analisada: "{palavra}"')
print(f'   Synsets encontrados: {len(synsets)}')

# Trabalhar com o primeiro synset
syn = synsets[0]
print(f'\n📌 Synset principal: {syn.name()}')
print(f'   Definição : {syn.definition()}')
print(f'   Sinônimos : {[l.name() for l in syn.lemmas()]}')

# Relações
print(f'\n🔗 Relações:')
print(f'   Hypernyms : {[h.name() for h in syn.hypernyms()]}')
print(f'   Hyponyms  : {[h.name() for h in syn.hyponyms()[:3]]} ...')

# Similaridade
university = wn.synset('university.n.01')
hospital   = wn.synset('hospital.n.01')
sim_univ   = syn.path_similarity(university)
sim_hosp   = syn.path_similarity(hospital)
print(f'\n📊 Similaridade semântica:')
print(f'   school × university : {sim_univ:.4f} 🟢')
print(f'   school × hospital   : {sim_hosp:.4f} 🟡')

# Português
lemas_pt = syn.lemma_names(lang='por')
print(f'\n🇧🇷 Equivalente em Português: {lemas_pt}')

print()
print('=' * 60)
print('  ✅ Todos os conceitos da Aula 1 demonstrados!')
print('=' * 60)
````
