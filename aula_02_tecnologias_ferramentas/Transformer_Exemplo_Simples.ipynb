# 🤖 Transformers na Prática — Exemplo Simples
### Usando o Hugging Face para tarefas reais em 5 minutos

---

Este notebook demonstra o Transformer da forma mais direta possível:  
**uma função, um modelo, um resultado.**

Nenhuma matemática. Nenhuma configuração complexa. Só PLN funcionando.

---

```python
# ─────────────────────────────────────────────────────────────
# PASSO 1 — Instalar e importar
# A biblioteca 'transformers' do Hugging Face dá acesso a
# centenas de modelos pré-treinados com uma linha de código
# ─────────────────────────────────────────────────────────────
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

print('✅ Pronto! Vamos usar Transformers.')
```

---
## 😊 Exemplo 1 — Analisar o sentimento de um texto

O modelo lê uma frase e decide se ela é **positiva** ou **negativa**.

```python
# ─────────────────────────────────────────────────────────────
# ANÁLISE DE SENTIMENTO
# pipeline() carrega automaticamente o melhor modelo
# para a tarefa pedida
# ─────────────────────────────────────────────────────────────

# Criar o analisador (baixa o modelo na primeira vez)
analisador = pipeline('sentiment-analysis')

# Testar com algumas frases
frases = [
    'I love this course, it is fantastic!',
    'This is the worst experience I have ever had.',
    'The weather today is okay, nothing special.',
]

print('Frase                                        Sentimento   Confiança')
print('-' * 65)

for frase in frases:
    resultado  = analisador(frase)[0]
    sentimento = resultado['label']
    confianca  = resultado['score']
    emoji      = '😊' if sentimento == 'POSITIVE' else '😞'

    print(f'{frase[:44]:<45} {emoji} {sentimento:<10} {confianca:.0%}')
```

---
## 🎭 Exemplo 2 — Adivinhar uma palavra escondida

O Transformer lê a frase com um `[MASK]` e **adivinha qual palavra falta**.  
Isso mostra que ele entende o contexto — não chuta aleatoriamente.

```python
# ─────────────────────────────────────────────────────────────
# FILL-MASK — o modelo completa a frase
# [MASK] marca o lugar onde a palavra foi removida
# ─────────────────────────────────────────────────────────────

adivinhador = pipeline('fill-mask', model='bert-base-uncased')

frase_com_mascara = 'The capital of France is [MASK].'

print(f'Frase: "{frase_com_mascara}"')
print()
print('Top 3 palavras previstas pelo BERT:')
print('-' * 40)

resultados = adivinhador(frase_com_mascara, top_k=3)

for i, r in enumerate(resultados):
    palavra    = r['token_str']
    confianca  = r['score']
    barra      = '█' * int(confianca * 30)
    print(f'{i+1}. "{palavra}"  {confianca:.1%}  {barra}')
```

---
## ❓ Exemplo 3 — Responder uma pergunta sobre um texto

Você dá um **contexto** e uma **pergunta**.  
O Transformer encontra a resposta dentro do texto — sem inventar nada.

```python
# ─────────────────────────────────────────────────────────────
# QUESTION ANSWERING
# O modelo lê o contexto e extrai a resposta da pergunta
# Ele NÃO inventa — só responde com o que está no texto
# ─────────────────────────────────────────────────────────────

respondedor = pipeline('question-answering')

contexto = '''
    Python was created by Guido van Rossum and first released in 1991.
    It is a high-level programming language known for its simple syntax.
    Python is widely used in data science, web development, and AI.
'''

perguntas = [
    'Who created Python?',
    'When was Python first released?',
    'What is Python used for?',
]

print('Pergunta                            Resposta              Confiança')
print('-' * 70)

for pergunta in perguntas:
    resultado  = respondedor(question=pergunta, context=contexto)
    resposta   = resultado['answer']
    confianca  = resultado['score']
    print(f'{pergunta:<36} {resposta:<22} {confianca:.0%}')
```

---
## 📋 Exemplo 4 — Classificar um texto em categorias

**Sem treinar nada.** Você define as categorias na hora  
e o modelo classifica o texto — isso é **zero-shot classification**.

```python
# ─────────────────────────────────────────────────────────────
# ZERO-SHOT CLASSIFICATION
# Classifica texto em categorias que você define agora
# O modelo nunca foi treinado com essas categorias!
# ─────────────────────────────────────────────────────────────

classificador = pipeline('zero-shot-classification',
                          model='facebook/bart-large-mnli')

texto     = 'The stock market crashed today as investors feared rising inflation.'
categorias = ['finance', 'sports', 'technology', 'politics', 'health']

resultado = classificador(texto, candidate_labels=categorias)

print(f'Texto: "{texto}"')
print()
print('Categoria         Confiança   Barra')
print('-' * 50)

for label, score in zip(resultado['labels'], resultado['scores']):
    barra   = '█' * int(score * 25) + '░' * (25 - int(score * 25))
    destaque = ' ← vencedor' if label == resultado['labels'][0] else ''
    print(f'{label:<18} {score:.0%}    {barra}{destaque}')
```

---
## 📝 Exemplo 5 — Resumir um texto longo

O modelo lê um parágrafo longo e gera um **resumo automático**.

```python
# ─────────────────────────────────────────────────────────────
# SUMMARIZATION — resumo automático
# max_length → tamanho máximo do resumo
# min_length → tamanho mínimo (evita resumos de 1 palavra)
# ─────────────────────────────────────────────────────────────

resumidor = pipeline('summarization', model='facebook/bart-large-cnn')

texto_longo = '''
    Artificial intelligence has made remarkable progress in recent years,
    particularly in the field of natural language processing. The introduction
    of transformer-based models like BERT and GPT has revolutionized how
    computers understand and generate human language. These models are trained
    on vast amounts of text data and can perform a wide range of tasks,
    from answering questions to translating languages and summarizing documents.
    Companies around the world are now integrating these technologies into
    their products and services, transforming industries such as healthcare,
    education, finance, and customer service.
'''

resumo = resumidor(texto_longo, max_length=60, min_length=20)[0]['summary_text']

print('📄 TEXTO ORIGINAL:')
print(texto_longo.strip())
print()
print('📋 RESUMO GERADO PELO TRANSFORMER:')
print(resumo)
```

---
## 🔁 Exemplo 6 — Traduzir português → inglês

Tradução automática usando um modelo Transformer treinado  
especificamente para o par de idiomas **PT → EN**.

```python
# ─────────────────────────────────────────────────────────────
# TRADUÇÃO — Português para Inglês
# Helsinki-NLP treinou modelos para centenas de pares de idiomas
# opus-mt-ORIGEM-DESTINO (pt = português, en = inglês)
# ─────────────────────────────────────────────────────────────

tradutor = pipeline('translation',
                     model='Helsinki-NLP/opus-mt-pt-en')

frases_pt = [
    'O processamento de linguagem natural é fascinante.',
    'Redes neurais aprendem padrões a partir de dados.',
    'O Brasil é o maior país da América do Sul.',
]

print('Português                                    → Inglês')
print('-' * 75)

for frase in frases_pt:
    traducao = tradutor(frase)[0]['translation_text']
    print(f'{frase:<44} → {traducao}')
```
