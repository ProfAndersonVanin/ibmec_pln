## 📚 2. Aula 1 — Representações Conceituais (NLTK / NetworkX)

````python
import nltk, networkx
print(f'NLTK     : {nltk.__version__}')
print(f'NetworkX : {networkx.__version__}')

# Baixar recursos essenciais do NLTK
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

from nltk.corpus import wordnet as wn
syns = wn.synsets('dog')
assert len(syns) > 0, '❌ WordNet não carregou'
print(f'WordNet OK — Synsets de "dog": {syns[0].name()}')
print('\n✅ Aula 1 OK')
````

## 🛠️ 3. Aula 2 — Ferramentas PLN (spaCy)

````python
import spacy
print(f'spaCy : {spacy.__version__}')

nlp_en = spacy.load('en_core_web_sm')
nlp_pt = spacy.load('pt_core_news_sm')

doc_en = nlp_en('Natural Language Processing is fascinating.')
doc_pt = nlp_pt('Processamento de Linguagem Natural é fascinante.')

print(f'Modelo EN carregado — tokens: {[t.text for t in doc_en]}')
print(f'Modelo PT carregado — tokens: {[t.text for t in doc_pt]}')
print('\n✅ Aula 2 — spaCy OK')
````

## 🧠 5. Aula 3 — Arquiteturas (PyTorch)

````python
import torch
print(f'PyTorch : {torch.__version__}')
print(f'CUDA disponível: {torch.cuda.is_available()}')

# Teste simples de LSTM
lstm = torch.nn.LSTM(input_size=10, hidden_size=20, num_layers=1, batch_first=True)
x = torch.randn(5, 3, 10)
out, _ = lstm(x)
assert out.shape == (5, 3, 20), '❌ LSTM com shape incorreto'
print(f'LSTM OK — output shape: {out.shape}')
print('\n✅ Aula 3 OK')
````

## 🤗 6. Aula 4 — Modelos Pré-Treinados (Transformers / Hugging Face)

````python
import transformers, datasets, tokenizers
print(f'Transformers : {transformers.__version__}')
print(f'Datasets     : {datasets.__version__}')
print(f'Tokenizers   : {tokenizers.__version__}')

# Teste com DistilBERT (versão leve — adequada para CPU)
from transformers import pipeline
print('\nCarregando DistilBERT (pode demorar na primeira vez)...')
classifier = pipeline(
    'sentiment-analysis',
    model='distilbert-base-uncased-finetuned-sst-2-english'
)
result = classifier('I love studying Natural Language Processing!')
print(f'Resultado: {result}')
assert result[0]['label'] == 'POSITIVE', '❌ Classificação inesperada'
print('\n✅ Aula 4 OK')
````

## 🖼️ 7. Aula 5 — Multimodal (Pillow / Sentence-Transformers)

````python
from PIL import Image
from sentence_transformers import SentenceTransformer
import PIL
print(f'Pillow              : {PIL.__version__}')

print('Carregando Sentence-Transformer...')
model = SentenceTransformer('all-MiniLM-L6-v2')
emb = model.encode(['PLN é incrível', 'Deep Learning é poderoso'])
assert emb.shape[0] == 2, '❌ Embeddings com shape incorreto'
print(f'Embeddings shape: {emb.shape}')
print('\n✅ Aula 5 OK')
````

## 📝 8. Aula 6 — Aplicações PLN (BLEU / ROUGE / TextBlob)

````python
import sacrebleu
from rouge_score import rouge_scorer
from textblob import TextBlob

# Teste BLEU
refs = [['The cat is on the mat.']]
hyp = ['The cat is on the mat.']
bleu = sacrebleu.corpus_bleu(hyp, refs)
print(f'sacrebleu : {sacrebleu.__version__} | BLEU perfeito: {bleu.score:.1f}')

# Teste ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1'], use_stemmer=True)
scores = scorer.score('The cat is on the mat.', 'The cat is on the mat.')
print(f'ROUGE-1 F1: {scores["rouge1"].fmeasure:.2f}')

# Teste Sentimento
blob = TextBlob('Natural Language Processing is amazing!')
print(f'TextBlob sentimento: {blob.sentiment.polarity:.2f}')
print('\n✅ Aula 6 OK')
````
